```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load ADSB-like data
file_path = '..\AircraftADSBData.xlsx'
data = pd.read_excel(file_path)
print("Data Loaded from Excel:")
```

```python
# Display first few rows
print("First few rows of the dataset:")
print(data.head())
```

```python
# Scatter pf Altitude vs Speed with color coding for different ICAO
plt.figure(figsize=(10, 6))
colors = plt.colormaps.get_cmap('tab10').resampled(len(data['ICAO'].unique()))
for i, icao in enumerate(data['ICAO'].unique()):
    subset = data[data['ICAO'] == icao]
    plt.scatter(subset['Ground Speed'], subset['Altitude'], color=colors(i), label=str(icao), alpha=0.6)
plt.title('Altitude vs Speed for each ICAO')
plt.xlabel('Speed (knots)')
plt.ylabel('Altitude (feet)')
plt.legend(title='ICAO')
plt.grid(True)
plt.show()
```
